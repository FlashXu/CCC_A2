```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

df = pd.read_csv('WholeData.csv')
df['TeenRatio'] = df.TeenNum / df.TotalNum
df = df[df.TweetNum < 150000]
df = df[df.TeenRatio > 0.1]
df.TweetNum = df.TweetNum / 10000

x = df.TeenRatio
y = df.TweetNum
plt.scatter(x, y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', linestyle=':')
plt.title('Teen Ratio & Tweeting Frequency (Corr = -0.26)')
plt.xlabel('Teen Ratio')
plt.ylabel('Number of Tweets x 10000')
plt.show()
corr, _ = pearsonr(x, y)
corr
```
